# Data Analysis

This dataset compiles a wide range of information aimed at understanding, analyzing, and predicting depression levels among students. It is designed for research in psychology, data science, and education, providing insights into factors that contribute to student mental health challenges and aiding in the design of early intervention strategies.

# Data Description

- Format: CSV (each row represents an individual student)
- Features:
    - ID: Unique identifier for each student
    - Demographics: Age, Gender, City
    - Academic Indicators: CGPA, Academic Pressure, Study Satisfaction
    - Lifestyle & Wellbeing: Sleep Duration, Dietary Habits, Work Pressure, Job Satisfaction, Work/Study Hours
    - Additional Factors: Profession, Degree, Financial Stress, Family History of Mental Illness, and whether the student has ever had suicidal thoughts
- Target Variable:
    - Depression_Status: A binary indicator (0/1 or Yes/No) that denotes whether a student is experiencing depression
  
# Key Highlights

Multifaceted Data: Integrates demographic, academic, and lifestyle factors to offer a comprehensive view of student wellbeing.
Ethical Considerations: Data collection adhered to strict ethical standards with an emphasis on privacy, informed consent, and anonymization.
Research & Practical Applications: Ideal for developing predictive models, conducting statistical analyses, and informing mental health intervention strategies in educational environments.
Usage & Potential Applications

Academic Research: Explore correlations between academic pressures and mental health trends.
Data Science Projects: Build predictive models to identify at-risk students based on various indicators.
Policy Making: Inform the development of targeted mental health support programs within academic institutions.

## Preprocessing Categorical Data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
drop_columns = []
```

```python
# Com base no estudo, o campo cidade não represensa quaisquer indício que resulte em uma forte correlação há casos de Depressão.
df.drop('City', axis=1, inplace=True)
```

---

```python
df['Gender'].unique()
```

```python
label_gender = LabelEncoder()

df['Gender_cat'] =  label_gender.fit_transform(df['Gender'])

drop_columns.append('Gender')

df[['Gender', 'Gender_cat']].head()
```

---

```python
df['Profession'].unique()
```

```python
df['Profession'] = df['Profession'].str.replace("'", "").str.strip()
```

```python
# Usando get_dummies
df = pd.get_dummies(df, columns=['Profession'])
```

```python
# Convertendo as colunas geradas de bool para int
df[df.columns[df.columns.str.contains('Profession')]] = df[df.columns[df.columns.str.contains('Profession')]].astype(float)
```

--- 

```python
df['Sleep Duration'].unique()
```

```python
(df['Sleep Duration'] == 'Others').sum()
```

```python
# Vemos que a Others não representa uma ordem que possa se encaixar ao restante, e representando apenas 18 registros podemos excluir.
df = df[df['Sleep Duration'] != 'Others']

drop_columns.append('Sleep Duration')

df['Sleep Duration'].unique()
```

```python
map_sleep_duration = {
    "'Less than 5 hours'": 0,
    "'5-6 hours'": 1,
    "'7-8 hours'": 2,
    "'More than 8 hours'": 3,
}

df['Sleep_Duration_cat'] = df['Sleep Duration'].map(map_sleep_duration)

df[['Sleep Duration', 'Sleep_Duration_cat']].head()
```

---

```python
df['Degree'].unique()
```

##### Em nossa perspectiva o Degree não possa ser uma grande influencia em casos de Depressão, mesmo que seja mínima e reconhecendo o poder computacional, é justo pensar que remover a coluna possar ser a melhor decisão

```python
df.drop('Degree', axis=1, inplace=True)
```

---

```python
df['Have you ever had suicidal thoughts ?'].unique()
```

```python
label_suicidal = LabelEncoder()
df['suicidal_thinking'] = label_suicidal.fit_transform(df['Have you ever had suicidal thoughts ?'])

drop_columns.append('Have you ever had suicidal thoughts ?')

df[['Have you ever had suicidal thoughts ?', 'suicidal_thinking']].head()
```

---

```python
df['Dietary Habits'].unique()
```

- Seguindo o mesmo contexto da duração de um sono

```python
(df['Dietary Habits'] == 'Others').sum()
```

```python
df = df[df['Dietary Habits'] != 'Others']

drop_columns.append('Dietary Habits')

map_dietary_habits = {
    'Unhealthy': 0,
    'Moderate': 1,
    'Healthy': 2
}

df['Dietary Habits_cat'] = df['Dietary Habits'].map(map_dietary_habits)
```

---

```python
df['Financial Stress'].unique()
```

```python
(df['Financial Stress'] == '?').sum()
```

```python
df = df[df['Financial Stress'] != '?']

df['Financial Stress'] = df['Financial Stress'].astype(float)
```

---

```python
df['Family History of Mental Illness'].unique()
```

```python
label_encoder_family_history = LabelEncoder()

drop_columns.append('Family History of Mental Illness')

df['Family_History_Of_Mental_Illness_cat'] =  label_encoder_family_history.fit_transform(df['Family History of Mental Illness'])
```

# Preprocessing 2

```python
df = df.astype(float)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('Depression', axis=1).values
y = df['Depression'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
```

```python
X_test_scaler = scaler.transform(X_test)
```

# Model - XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(objective='binary:logistic')
```

```python
xgb.fit(X_train_scaler, y_train)
```

```python
pred = xgb.predict(X_test_scaler)
```

# Evaluation

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, pred))
```

```python
print(accuracy_score(y_test, pred))
```

```python
cm = confusion_matrix(y_test, pred)

cm_disp = ConfusionMatrixDisplay(confusion_matrix=cm)

cm_disp.plot()
```

# Oversampling - Smote

```python
sns.countplot(x='Depression', data=df)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
smote = SMOTE(sampling_strategy='minority', random_state=42, k_neighbors=5)
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print('Resampled dataset shape %s' % Counter(y_train_smote))
```

```python
scaler = MinMaxScaler()
```

```python
X_train_smote_scaler = scaler.fit_transform(X_train_smote)
X_test_scaler = scaler.transform(X_test)
```

```python
X_test_scaler.min(), X_test_scaler.max()
```

```python
xgb = XGBClassifier(objective='binary:logistic')
```

```python
xgb.fit(X_train_smote_scaler, y_train_smote)
```

```python
pred = xgb.predict(X_test_scaler)
```

```python
print(classification_report(y_test, pred))
```

# Tuning - GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb = XGBClassifier(objective='binary:logistic', eval_metric='logloss')

param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.0001, 0.001, 0.01, 0.1, 0.2],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1],
    'gamma': [0, 1],
    'reg_alpha': [0, 0.1],
    'reg_lambda': [1, 10],
}

grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, n_jobs=-1, cv=5, verbose=1, scoring='accuracy')

grid_search.fit(X_train_smote_scaler, y=y_train_smote)
```

```python
print("Melhores hiperparâmetros encontrados:")
print(grid_search.best_params_)

print("="*100)

print("Melhor estimador encontrado:")
print(grid_search.best_estimator_)
best_param_grid = grid_search.best_estimator_
```

```python
xgb_gridsearch = best_param_grid

xgb_gridsearch.fit(X_train_smote_scaler, y_train_smote)
```

```python
pred = xgb_gridsearch.predict(X_test_scaler)
```

```python
print(classification_report(y_test, pred))
```

```python
print(round(accuracy_score(y_test, pred) * 100, 2))
```

```python
cm = confusion_matrix(y_test, pred)

cm_disp = ConfusionMatrixDisplay(confusion_matrix=cm)

cm_disp.plot()
```

# Model - MLP

```python
X_train_smote_scaler.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Input(shape=(27,)))

model.add(Dense(units=54, kernel_regularizer=l2(), activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(units=81, kernel_regularizer=l2(), activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))

model.add(Dense(units=54, kernel_regularizer=l2(), activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(units=27, kernel_regularizer=l2(), activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))

model.add(Dense(units=14, kernel_regularizer=l2(), activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))

# model.add(Dense(units=7, kernel_regularizer=l2(), activation='relu'))
# model.add(BatchNormalization())
# model.add(Dropout(0.1))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
early = EarlyStopping(monitor='val_loss', patience=10, mode='min')
```

```python
model.fit(X_train_smote_scaler, y_train_smote, batch_size=32, epochs=50, callbacks=[early], validation_data=(X_test_scaler, y_test), shuffle=True)
```

---

```python
model2 = Sequential([
    Input(shape=(27,)),
    Dense(64, activation='relu', kernel_regularizer=l2()),
    BatchNormalization(),
    Dropout(0.3),
    
    Dense(32, activation='relu', kernel_regularizer=l2()),
    BatchNormalization(),
    Dropout(0.3),
    
    Dense(1, activation='sigmoid')
])

model2.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```
